```python

import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob

import librosa
import librosa.display
import IPython.display as ipd

from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
pd.Series(y_trimmed).plot(figsize=(10, 5),
                  lw=1,
                  title='Raw Audio Trimmed Example',
                 color=color_pal[3])
plt.show()
```

```python


import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob

import librosa
import librosa.display
import IPython.display as ipd

from itertools import cycle

# Trim leading/lagging silence
y_trimmed, _ = librosa.effects.trim(y, top_db=20)

# Define start and end time for segmentation (in seconds)
start_time = 1.0
end_time = 2.0

# Convert time to samples
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)

# Segment the audio based on time
y_segmented = y[start_sample:end_sample]

# Plot the segmented audio
plt.figure(figsize=(10, 5))
plt.plot(y_segmented, color='blue', label='Trimmed Audio')
plt.title('Trimmeded Audio ')
plt.legend()
plt.show()
```

```python

# Play the trimmed audio
ipd.Audio(y_segmented, rate=sr)
```
